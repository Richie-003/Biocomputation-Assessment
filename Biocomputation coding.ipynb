```python
# Function to calculate mutation type and location
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from Bio.Seq import Seq
# Assuming the DataFrame is already created and named 'mutation_data'
# Let's analyze the data in detail

# Function to determine mutation type and location
def analyze_mutation(wt_seq, mut_seq):
    """
    Analyze differences between wild-type and mutant sequences
    Returns: mutation type, location (promoter/coding), specific change
    """
    if len(wt_seq) > len(mut_seq):
        mutation_type = "Deletion"
        # Find where the sequences differ
        for i in range(min(len(wt_seq), len(mut_seq))):
            if wt_seq[i] != mut_seq[i]:
                break
        location = "Promoter" if i < 1000 else "Coding"
        deleted_bases = len(wt_seq) - len(mut_seq)
        specific_change = f"Deletion of {deleted_bases} bp starting at position {i+1}"
        
    elif len(mut_seq) > len(wt_seq):
        mutation_type = "Insertion"
        # Find where the sequences differ
        for i in range(min(len(wt_seq), len(mut_seq))):
            if wt_seq[i] != mut_seq[i]:
                break
        location = "Promoter" if i < 1000 else "Coding"
        inserted_bases = len(mut_seq) - len(wt_seq)
        specific_change = f"Insertion of {inserted_bases} bp starting at position {i+1}"
        
    else:  # Same length, must be substitution
        mutation_type = "Substitution"
        differences = []
        for i in range(len(wt_seq)):
            if wt_seq[i] != mut_seq[i]:
                differences.append(i)
        
        if len(differences) == 1:
            i = differences[0]
            location = "Promoter" if i < 1000 else "Coding"
            specific_change = f"{wt_seq[i]} to {mut_seq[i]} at position {i+1}"
        else:
            # Multiple substitutions
            promoter_changes = sum(1 for i in differences if i < 1000)
            coding_changes = len(differences) - promoter_changes
            
            if promoter_changes > coding_changes:
                location = "Promoter"
            elif coding_changes > promoter_changes:
                location = "Coding"
            else:
                location = "Both"
            
            specific_change = f"{len(differences)} substitutions"
    
    return mutation_type, location, specific_change

# Calculate effects on expression and viability
def calculate_effects(row):
    """
    Calculate percent changes in mRNA, protein expression, and cell viability
    """
    mrna_wt_avg = (row['mRNA.Expression.WT.Rep1'] + row['mRNA.Expression.WT.Rep2'] + row['mRNA.Expression.WT.Rep3']) / 3
    mrna_mut_avg = (row['mRNA.Expression.Mut.Rep1'] + row['mRNA.Expression.Mut.Rep2'] + row['mRNA.Expression.Mut.Rep3']) / 3
    mrna_change = ((mrna_mut_avg - mrna_wt_avg) / mrna_wt_avg) * 100
    
    protein_wt_avg = (row['Protein.Expression.WT.Rep1'] + row['Protein.Expression.WT.Rep2'] + row['Protein.Expression.WT.Rep3']) / 3
    protein_mut_avg = (row['Protein.Expression.Mut.Rep1'] + row['Protein.Expression.Mut.Rep2'] + row['Protein.Expression.Mut.Rep3']) / 3
    protein_change = ((protein_mut_avg - protein_wt_avg) / protein_wt_avg) * 100
    
    viability_wt_avg = (row['CellViability.WT.Rep1'] + row['CellViability.WT.Rep2'] + row['CellViability.WT.Rep3']) / 3
    viability_mut_avg = (row['CellViability.Mut.Rep1'] + row['CellViability.Mut.Rep2'] + row['CellViability.Mut.Rep3']) / 3
    viability_change = ((viability_mut_avg - viability_wt_avg) / viability_wt_avg) * 100
    
    return pd.Series([mrna_change, protein_change, viability_change, 
                     mrna_wt_avg, mrna_mut_avg, protein_wt_avg, protein_mut_avg, 
                     viability_wt_avg, viability_mut_avg])

# Check if coding mutations affect amino acid sequence
def analyze_protein_effect(wt_seq, mut_seq):
    """
    Determine if mutation causes amino acid change and what type
    """
    # Extract coding regions (after 1000bp promoter)
    wt_coding = wt_seq[1000:]
    mut_coding = mut_seq[1000:]
    
    # If the mutation is only in the promoter
    if wt_coding == mut_coding:
        return "No change (promoter mutation only)"
    
    # Try to translate both sequences
    try:
        # Make sure we have complete codons
        wt_coding_adjusted = wt_coding[:len(wt_coding) - (len(wt_coding) % 3)]
        mut_coding_adjusted = mut_coding[:len(mut_coding) - (len(mut_coding) % 3)]
        
        wt_protein = str(Seq(wt_coding_adjusted).translate())
        mut_protein = str(Seq(mut_coding_adjusted).translate())
        
        if wt_protein == mut_protein:
            return "Synonymous mutation"
        elif '*' in mut_protein and '*' not in wt_protein[:len(mut_protein)]:
            return "Premature stop codon"
        elif len(mut_protein) < len(wt_protein):
            return "Truncated protein"
        elif len(mut_protein) > len(wt_protein):
            return "Extended protein"
        else:
            # Count amino acid changes
            aa_changes = sum(1 for a, b in zip(wt_protein, mut_protein) if a != b)
            return f"{aa_changes} amino acid change(s)"
    except:
        return "Unable to determine protein effect"

# Let's assume we have the data in a DataFrame
# Apply the analysis functions to each row
results = []

# For demonstration, let's create a simplified version of what we might process
# In a real scenario, you would loop through your actual mutation_data DataFrame
for _, row in entire_data.iterrows():
    gene_name = row['Gene']
    wt_seq = row['WildType.Sequence']
    mut_seq = row['Mutant.Sequence']
    
    # Analyze mutation
    mutation_type, location, specific_change = analyze_mutation(wt_seq, mut_seq)
    
    # Calculate expression and viability effects
    effects = calculate_effects(row)
    
    # Analyze protein effect
    protein_effect = analyze_protein_effect(wt_seq, mut_seq)
    
    # Compile results
    result = {
        'Gene': gene_name,
        'Mutation_Type': mutation_type,
        'Mutation_Location': location,
        'Specific_Change': specific_change,
        'mRNA_Change_%': effects[0],
        'Protein_Change_%': effects[1],
        'Viability_Change_%': effects[2],
        'mRNA_WT_Avg': effects[3],
        'mRNA_Mut_Avg': effects[4],
        'Protein_WT_Avg': effects[5],
        'Protein_Mut_Avg': effects[6],
        'Viability_WT_Avg': effects[7],
        'Viability_Mut_Avg': effects[8],
        'Protein_Effect': protein_effect
    }
    results.append(result)

# Create final DataFrame
analysis_results = pd.DataFrame(results)

# Create a prioritization score based on the magnitude of effects
analysis_results['Priority_Score'] = (
    analysis_results['Viability_Change_%']+  # Viability is most important
    analysis_results['Protein_Change_%']+    # Protein expression second
    analysis_results['mRNA_Change_%']        # mRNA expression third
)

# Prioritize results where viability increases
analysis_results.loc[analysis_results['Viability_Change_%'] > 0, 'Priority_Score'] *= 1.5

# Get top 5 genes based on priority score
top_genes = analysis_results.sort_values('Priority_Score', ascending=False).head(5)
print("Top 5 Genes for Prioritization:")
print(top_genes[['Gene', 'Mutation_Type', 'Mutation_Location', 'Viability_Change_%','mRNA_Change_%','Protein_Change_%', 'Priority_Score']])
```

```python
# Create visualizations
plt.figure(figsize=(12, 8))

# Boxplot of changes by mutation type
plt.subplot(2, 2, 1)
sns.boxplot(x='Mutation_Type', y='Viability_Change_%', data=top_genes)
plt.title('Effect of Mutation Type on Cell Viability')
plt.xlabel('Mutation Type')
plt.ylabel('% Change in Cell Viability')

# Scatter plot of mRNA vs protein changes
plt.subplot(2, 2, 2)
sns.scatterplot(x='mRNA_Change_%', y='Protein_Change_%', 
                hue='Viability_Change_%', data=top_genes)
plt.title('Relationship Between mRNA and Protein Expression Changes')
plt.xlabel('% Change in mRNA Expression')
plt.ylabel('% Change in Protein Expression')

# KDE plot of viability changes by mutation location
plt.subplot(2, 2, 3)
sns.kdeplot(data=top_genes, x='Viability_Change_%', hue='Mutation_Location')
plt.title('Distribution of Viability Changes by Mutation Location')
plt.xlabel('% Change in Cell Viability')
plt.ylabel('Density')

plt.tight_layout()
plt.savefig('mutation_analysis_plots.png', dpi=300)
plt.show()
```
